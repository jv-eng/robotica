```python
#escenario esfera centrada
vect = [24.07852370172136, 37.519576746156645, 78.34269453599595, 109.40210646788142, 122.76390489310184, 140.38148141875064, 194.64861920894316, 215.7735577859291]
datos = np.arange(0, 256, 0.1)

plt.subplot(3, 1, 1)
plt.plot(datos, fuzz.trapmf(datos, [0, 0, vect[0], vect[1]]), 'b', linewidth=1.5, label='muy izq')
plt.plot(datos, fuzz.trapmf(datos, [vect[0], vect[1], vect[2], vect[3]]), 'g', linewidth=1.5, label='izq')
plt.plot(datos, fuzz.trapmf(datos, [vect[2], vect[3], vect[4], vect[5]]), 'k', linewidth=1.5, label='centrado')
plt.plot(datos, fuzz.trapmf(datos, [vect[4], vect[5], vect[6], vect[7]]), 'r', linewidth=1.5, label='dcha')
plt.plot(datos, fuzz.trapmf(datos, [vect[6], vect[7], 256, 256]), 'm', linewidth=1.5, label='muy dcha')
plt.title('Error actual')
plt.xlabel('Valor')
plt.ylabel('Membresía')
plt.legend()
plt.tight_layout()
plt.show()
```
